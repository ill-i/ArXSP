```python
import sys
from PyQt5.QtWidgets import QApplication, QFileDialog
from astropy.io import fits
from HeaderEditWindow import FileHeaderEditor  # <-- Исправленный импорт

class ArxData:
    def __init__(self, header):
        self._header = header

    def get_header(self):
        return self._header

if __name__ == '__main__':
    app = QApplication(sys.argv)

    options = QFileDialog.Options()
    filepath, _ = QFileDialog.getOpenFileName(
        None,
        "Выберите FITS-файл",
        "",
        "FITS-файлы (*.fits *.fit);;Все файлы (*)",
        options=options
    )

    if filepath:
        hdu = fits.open(filepath)
        header = hdu[0].header
        hdu.close()

        arx_data = ArxData(header)
        window = FileHeaderEditor(arx_data=arx_data)  # <-- правильный вызов

        def print_saved_header(hdr):
            print("Заголовок сохранён:")
            print(hdr)

        window.headerSaved.connect(print_saved_header)
        window.show()

        sys.exit(app.exec_())
```
